```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/card_transdata.csv")
df.head()
```

```python
print(df.info())
```

```python
#Paso a factor la columna de clase
df['fraud'] = df['fraud'].astype('category')
print(df['fraud'].dtype)
```

```python
# Establecer la semilla para reproducibilidad
np.random.seed(30162)
```

```python
from sklearn.model_selection import train_test_split

# 'fraud' es la columna a predecir
X = df.drop(columns=['fraud'])  # Características
y = df['fraud']  # Variable objetivo

# Dividir el conjunto de datos en conjuntos de entrenamiento y prueba
# Usamos stratify=y para mantener la misma proporción de clases en el conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=30162)

# Imprimir el número de filas en el conjunto de entrenamiento
print(len(X_train))
```

```python
print(X_train.info())
```

```python
print(X_test.info())
```

# Algoritmo SVM

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# Establecer la semilla para reproducibilidad
np.random.seed(30162)
```

```python
# Crear y entrenar el modelo SVM
svm_model = SVC(kernel='rbf', C=10, gamma=0.01)
svm_model.fit(X_train, y_train)
```

```python
y_pred = svm_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.metrics import confusion_matrix
# Calcular la matriz de confusión
cm = confusion_matrix(y_test, y_pred)
print("Matriz de Confusión:")
print(cm)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

# Calcular métricas de evaluación
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# Imprimir métricas
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)

# Imprimir un informe completo que incluye estas métricas para cada clase
print("Informe de Clasificación:")
print(classification_report(y_test, y_pred))
```

```python
#Guardo el modelo
from joblib import dump

# Guardar el modelo entrenado
dump(svm_model, '/content/svmodel.joblib')
```

# Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import StratifiedKFold, train_test_split
from sklearn.metrics import confusion_matrix
```

```python
# Definir las características y la variable objetivo para el conjunto de entrenamiento
X_trainGB = X_train[['distance_from_home', 'distance_from_last_transaction', 'ratio_to_median_purchase_price',
                     'repeat_retailer', 'used_chip', 'used_pin_number', 'online_order']]
y_trainGB = y_train
```

```python
# Definir la estrategia de validación cruzada
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=30162)

# Crear el modelo Gradient Boosting
boostModel = GradientBoostingClassifier(n_estimators=1000, random_state=30162)

# Entrenar el modelo Gradient Boosting
boostModel.fit(X_trainGB, y_trainGB)
```

```python
# Realizar predicciones en el conjunto de prueba
y_pred_testGBo = boostModel.predict(X_test)
```

```python
# Calcular la matriz de confusión
cm = confusion_matrix(y_test, y_pred_testGBo)
print("Matriz de Confusión:")
print(cm)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

# Calcular métricas de evaluación
accuracy = accuracy_score(y_test, y_pred_testGBo)
precision = precision_score(y_test, y_pred_testGBo)
recall = recall_score(y_test, y_pred_testGBo)
f1 = f1_score(y_test, y_pred_testGBo)

# Imprimir métricas
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)

# imprimir un informe completo que incluye estas métricas para cada clase
print("Informe de Clasificación:")
print(classification_report(y_test, y_pred_testGBo))
```

```python
#Guardo el modelo
from joblib import dump

# Guardar el modelo entrenado
dump(boostModel, '/content/GboostModel.joblib')
```
