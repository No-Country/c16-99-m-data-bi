```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/card_transdata.csv")
df.head()
```

```python
print(df.info())
```

```python
#Paso a factor la columna de clase
df['fraud'] = df['fraud'].astype('category')
print(df['fraud'].dtype)
```

```python
# Establecer la semilla para reproducibilidad
np.random.seed(30162)
```

```python
from sklearn.model_selection import train_test_split

# 'fraud' es la columna a predecir
X = df.drop(columns=['fraud'])  # Caracter√≠sticas
y = df['fraud']  # Variable objetivo

# Dividir el conjunto de datos en conjuntos de entrenamiento y prueba
# Usamos stratify=y para mantener la misma proporci√≥n de clases en el conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=30162)

# Imprimir el n√∫mero de filas en el conjunto de entrenamiento
print(len(X_train))
```

```python
print(X_train.info())
```

```python
print(X_test.info())
```

# Algoritmo SVM

```python
import sklearn
print(sklearn.__version__)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# Establecer la semilla para reproducibilidad
np.random.seed(30162)
```

```python
# Crear y entrenar el modelo SVM
svm_model = SVC(kernel='rbf', C=10, gamma=0.01)
svm_model.fit(X_train, y_train)
```

```python
y_pred = svm_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.metrics import confusion_matrix
# Calcular la matriz de confusi√≥n
cm = confusion_matrix(y_test, y_pred)
print("Matriz de Confusi√≥n:")
print(cm)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, balanced_accuracy_score, classification_report

# Calculo sensibilidad (recall)
sensitivity = recall_score(y_test, y_pred)

# Calculo la especificidad
specificity = cm[0, 0] / (cm[0, 0] + cm[0, 1])

# Calcular puntaje balanceado (balanced score)
balanced_score = balanced_accuracy_score(y_test, y_pred)

print("Sensibilidad (Recall):", sensitivity)
print("Especificidad:", specificity)
print("Puntaje balanceado:", balanced_score)
```

```python
import pickle

# Guardar el modelo entrenado con pickle
with open('/content/svm_model.pkl', 'wb') as f:
    pickle.dump(svm_model, f)
```
